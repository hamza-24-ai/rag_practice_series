```python

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_file = "data/who_substances_surveillance.pdf"

document_load = PyPDFLoader(
    file_path = text_file
)

document = document_load.load()

print(f"Total Documents : {len(document)}")
```

```python
print(f"Document Metadata : {document[0].metadata}")
```

```python
pdfData = RecursiveCharacterTextSplitter(
    chunk_size = 1200,
    chunk_overlap = 300
)

chunks = pdfData.split_documents(document)
print(f" Total Chunks : {len(chunks)}")
```

```python
sum = 0
for i, chunk in enumerate(chunks):
    print(f"Chunk Number : {i + 1}")
    print(f" Chunk Character Length : {len(chunk.page_content)}")
    sum += len(chunk.page_content)
    print("="*60)

print(f"Total Characters : {sum}")
```

```python
for i,chunk in enumerate(chunks[:2]):
    print("="*100)
    print(f" Chunk Number : {i+1}")
    print(f"Chunk Data :\n {chunk.page_content}")

print(type(chunks))
print(type(chunks[0]))
```

```python
#  Embedding System 
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("write your key here")
```

```python
text = [
    doc.page_content for doc in chunks
]

embedding = model.encode(text)

print(f"Length of embeddings : {len(embedding)}")
print(f"{embedding.shape[0]} Sentences each with {embedding.shape[1]} dimensions")
```

```python
print(f"First Embedding {embedding[0]}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_metrics = cosine_similarity(embedding)

print(f"Similarity chunk 0 and chunk 1 : {similarity_metrics[6][43]:.4f}")
```

```python
query = [
    "What is recommended for surveillance at 36 CCD"
]

query_embedding = model.encode(query)
```

```python
score  = cosine_similarity(
    query_embedding,
    embedding
)
print(score.shape)
```

```python
best_index = score.argmax()

print(best_index)
print(text[best_index])
print(f"Score : {score[0][best_index]}")
```
