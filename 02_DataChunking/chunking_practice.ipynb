```python

from langchain_community.document_loaders import TextLoader

def Loader(file_path):
    text_data = TextLoader( file_path = file_path )

    text_load = text_data.load()

    print(f"Text Document Length {len(text_load)}")
    print(f"Text data Content 300 characters {text_load[0].page_content [:300]}")

    print(f"Text metdata  {text_load[0].metadata}")

    print("="*90)
    len_data = text_load[0].page_content
    print(F"Total length of data : {len(len_data)}")
    start = 0;
    end = chunk_size = 80
    overlap = 20
    step = chunk_size - overlap
    chunk_number = 1

    while (start < len(len_data)) :
        print(f"\nchunk : {start} - {min(end,len(len_data))}")
        print(f"\nChunk_Number {chunk_number} \n {len_data[start:end]}")
        start += step
        end += step
        chunk_number +=1
        

text_path = "data/nlp_article.txt"
Loader(text_path)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader

text = TextLoader(file_path = "data/nlp_article.txt")
load = text.load()


text_split = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 200
)

chunks = text_split.split_documents(load)
print("="*60)

# Retrieve first 3 chunks
for i,chunk in enumerate(chunks[:3]):
    print(f"Chunk no {i + 1}")
    print(f"Chunk Length {len(chunk.page_content)}")
    print(f"Chunk data \n {chunk.page_content}")
    print(f"Meta Data \n {chunk.metadata}")
    print("="*60)

print(f"Total Number of chunks {len(chunks)}")
chunk_1 = chunks[0].page_content
chunk_2 = chunks[1].page_content

actual_overlap = 0
for size in range(min(300,len(chunk_2)), 0, -1):
    
    if chunk_1.endswith(chunk_2[:size]):
        actual_overlap = size
        break
print(f"Chunk 1 Length {len(chunk_1)}")
print(f"chunk 2 Length {len(chunk_2)}")

if actual_overlap > 0:
    print(" These Line hace showed bethween 2 chunks ")
    print(chunk_2[:actual_overlap])
else:
    print("No Overlap occur")
    print(actual_overlap)
    
```
