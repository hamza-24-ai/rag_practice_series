```python
description = [movie["description"] for movie in movies]
```

```python
movie_embedding = model.encode(description)

print(f"Prepare {len(movies)} in embeddings")
print(f"Data is {movie_embedding.shape[0]}  rows and {movie_embedding.shape[1]} dimensions")
```

```python
for i, data in enumerate(movies):
    print(f"{data["title"] :<25} -> {movie_embedding[i][0]:.3f}, {movie_embedding[i][1]:.3f}, {movie_embedding[i][2]:.3f} ....")
```

```python
from qdrant_client.models import PointStruct

points = []

for i, movie in enumerate(movies):
    point = PointStruct(
        id = movie["id"],
        vector = movie_embedding[i].tolist(),
        payload={
            "title":movie["title"],
            "year":movie["year"],
            "genre":movie["genre"],
            "description":movie["description"]
        }
    )
    points.append(point)

client.upsert(
    collection_name="movies",
    points=points
)

client.create_payload_index(
    collection_name="movies",
    field_name="genre",
    field_schema="keyword"
)

client.create_payload_index(
    collection_name="movies",
    field_name="year",
    field_schema="integer"
)

info = client.get_collection("movies")

print("Data is inserted Successfully")
print(f"Total Points : {info.points_count}")
    
```

```python
query = "A funny comedy about friends"
query_embedding = model.encode(query)

results = client.query_points(
    collection_name = "movies",
    query = query_embedding.tolist(),
    limit=3
).points

for index, movie in enumerate(results):
    print(f" Movie index {index + 1} \n Movie Title : {movie.payload["title"]} \n Movie Genre : {movie.payload["genre"]} \n Movie Description : {movie.payload["description"]}")
    print(f" Movie Score Value : {movie.score :.4f}")
    print("="*70)
```
